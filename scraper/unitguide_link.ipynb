```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Read the data from folder data/unit_info
unit_info = pd.read_csv('../data/unit_info.csv')
unit_info.head()
```

```python
# New df with unit_code, unit_name, unitguide_link
df = pd.DataFrame(columns=['unit_code', 'unit_name', 'unitguide_link', 'unitguide_year'])
df = pd.read_csv('../data/unitguide_link.csv')
count = 0
# Loop through the df and get the unit_code
for index, row in unit_info.iterrows():
    # Debug with count
    # if count == 10:
    #     break

    # Skip to index 1000
    if index < 12:
        continue
    unit_code = row['unit_code']
    unit_name = row['unit_name']
    url = f"https://unitguides.mq.edu.au/units/archive_search?query={unit_code}&year=2024"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    # Print for debugging
    print("Unit code: ", unit_code)
    print("Current index: ", index)

    # Get the first unit-guide-year class to see if there's a class for latest year
    if not soup.find('div', class_='unit-guide-year'):
        url = f"https://unitguides.mq.edu.au/units/archive_search?query={unit_code}&year="
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        if not soup.find('div', class_='unit-guide-year'):
            print("Not found classes for: ", unit_code)
            continue
    unitguide_year = soup.find('div', class_='unit-guide-year').find('h3').text
    unitguide_link = "https://unitguides.mq.edu.au" + soup.find('a', class_='unit-guide-list-item').get('href')
    df = pd.concat([df, pd.DataFrame([[unit_code, unit_name, unitguide_link, unitguide_year]], columns=['unit_code', 'unit_name', 'unitguide_link', 'unitguide_year'])], ignore_index=True)
    count += 1
# Save to a csv name unitguide_link.csv
df.to_csv('../data/unitguide_link.csv', index=False)
```
