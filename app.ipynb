```python
import pandas as pd

df = pd.read_excel("final_merged_data.xlsx")
```

```python
prompt = input().upper()
print(prompt)
filtered_df = df[df['Course Code'] == prompt]
# Select the desired columns
selected_columns = filtered_df[['Course Code', 'Major', 'Course Name', 'Period', 'reviews', 'stars', 'Links', 'Prerequisites', 'Corequisites', 'Co-badged Units', 'Description', 'all_final', 'all_presentation', 'Course Name_y', 'Tasks', 'Date']]

# Print the selected columns neatly
print(selected_columns.to_string(index=False))
```

```python
from tabulate import tabulate
import pandas as pd

# Filter the DataFrame based on "Course Code" value
filtered_df = df[df['Course Code'] == 'COMP2200']

# Select the desired columns
selected_columns = filtered_df[['Course Code', 'Major', 'Course Name', 'Period', 'reviews', 'stars', 'Links', 'Prerequisites', 'Corequisites', 'Co-badged Units', 'Description', 'all_final', 'all_presentation', 'Course Name_y', 'Tasks', 'Date']]

# Transpose the DataFrame
transposed_df = selected_columns.T

# Convert transposed DataFrame to tabular format
table = tabulate(transposed_df, headers='keys', tablefmt='fancy_grid')

# Print the formatted table
print(table)
print()  # Extra space below the table
```

```python
import streamlit as st

# Create a Streamlit app
st.title("Course Information Lookup")

# Input field for course name
course_name = st.text_input("Enter the Course Name:")

# Search and display course information
if st.button("Search"):
    # Filter the DataFrame based on the entered course name
    filtered_df = df[df['Course Name'] == course_name]

    # Display the filtered results
    if not filtered_df.empty:
        st.subheader("Course Information:")
        st.write(filtered_df)
    else:
        st.write("No matching course found.")

```
